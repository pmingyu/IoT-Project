```python
me1 = [i for i in df.columns if i.startswith('ME1')]
# me1 = [i for i in me1 if not i.endswith('ECC')]

to_exclude = ['ME1_FO_FLOW_HOUR_INLET',
            'ME1_FO_DENSITY_INLET',
            'ME1_FO_TEMP_INLET',
            'ME1_FO_TOTALIZER_INLET']

me1 = [i for i in me1 if i not in to_exclude]
me2 = ['SPEED_VG','lon','lat','TIME_STAMP']

df2=pd.concat([df[me1], df[me2]], axis=1)
df2 = df2.loc[:,~df2 .columns.duplicated()]

# Filter Value
df3=df2[(df2['ME1_RPM']>0) & (df2['ME1_SCAV_AIR_PRESS']>0) & (df2['ME1_LO_INLET_PRESS']>0) & (df2['ME1_FOC']>0)]
df3=df3[(df3['ME1_RPM']>df3['ME1_RPM'].describe()['75%'])  & (df3['SPEED_VG'] > 14)]

# Aggregating Cylinders Variable (Averaging)
pco=[i for i in df3.columns if 'PCO' in i]
cfw=[i for i in df3.columns if 'CFW' in i]
exh=[i for i in df3.columns if 'EXH' in i]
df3['ME1_CYL_PCO_OUTLET_TEMP']=df3[pco].mean(axis=1)
df3['ME1_CYL_CFW_OUTLET_TEMP']=df3[cfw].mean(axis=1)
df3['ME1_CYL_EXH_GAS_OUTLET_TEMP']=df3[exh].mean(axis=1)
df3 = df3.drop(pco+cfw+exh,axis=1)

df4 = df3.drop(['SPEED_VG','lon','lat','TIME_STAMP'],axis=1).copy()
```

총 59개의 ME1 컬럼 중에서
- 'ME1_FO_FLOW_HOUR_INLET' : 시간당 연료유 소비율
- 'ME1_FO_DENSITY_INLET' : 연료유 밀도
- 'ME1_FO_TEMP_INLET' : 연료유 온도
- 'ME1_FO_TOTALIZER_INLET' : 연료유 누적 소비량

**위 4개 컬럼은 중복된 컬럼이 존재하거나 필요 없는 컬럼이므로 제거**



**데이터 필터링**
- 'ME1_RPM' : 분당 엔진 회전수(RPM)
- 'ME1_SCAV_AIR_PRESS' : 공기 소거 압력
- 'ME1_LO_INLET_PRESS' : 윤활유 주입구 압력
- 'ME1_FOC' : 연료유 소비율

**위 4개의 컬럼이 0보다 큰(운항 중인 데이터)**
**또한 데이터 중 RPM의 75% 이상이며 속력이 14노트 이상인 값(엔진에 부하를 많이 주는 데이터)만을 필터링**


**컬럼 추가 전처리**
- 'ME1_CYL1-5_PCO_OUTLET_TEMP' : 1~5번 실린더 피스톤 냉각 오일의 출구 온도
- 'ME1_CYL1-5_CFW_OUTLET_TEMP' : 1~5번 실린더 블록 냉각수 출구 온도
- 'ME1_CYL1-5_EXH_GAS_OUTLET_TEMP' : 1~5번 실린더 배기 가스 배출구 온도

**위 3개 컬럼은 ME1 컬럼의 갯수가 매우 많으므로 1~5번 실린더의 각 PCO, CFW, EXH 컬럼 별 평균값으로 컬럼 병합**

**이후 진행 방향**
- df4를 Isolation Forest input하여 각 데이터별 ISO_SCORE 생성
- ISO_SCORE를 기준으로 히스토그램을 생성하여 이상치로 판단할 ISO_SCORE threshold 정의
- threshold 이하의 데이터들을 이상치로 정의
- Isolation Forest의 데이터를 SHAP에 input하여 각 컬럼 별 SHAP_VALUE 생성
- 정상데이터와 이상치데이터의 SHAP_VALUE 비교
- 이상치데이터의 각 컬럼 별 SHAP_VALUE만을 활용하여 Clustering 분석 진행
- 클러스터 별 특징 정의
- 원래의 df4의 컬럼 + ISO_SCORE + SHAP_VALUE + Cluster + TIME_STAMP + lat + lon 컬럼을 가진 datamart 생성
- datamart를 활용하여 Dashboard 생성
- Dashboard에 real-time을 구현할지 말지는 추후에 논의

## Isolation Forest

```python
iso = IsolationForest(random_state=4)
iso.fit(df4)
iso_score=iso.score_samples(df4)
```

```python
len(iso_score)
```

```python
plt.rcParams["figure.figsize"] = (12,9)
sns.distplot(iso_score)
plt.show()
```

## SHAP

```python
# SHAP Explainer
explainer = shap.TreeExplainer(iso,data=df4)
#Calculate shap with shap_values method
shap_values = explainer.shap_values(df4,check_additivity=True)
dfShap = pd.DataFrame(shap_values,columns=df4.columns,index=df4.index)

dfShap_temp = dfShap[:].copy()
dfShap_temp['ISO_SCORE'] = iso_score
```
